```python
#visualizing crime rates
import plotly.graph_objects as go
year = crime[crime.DISTRICT == 'Total']
year = year.groupby(['Year'])['Murder', 'Rape',
       'Kidnapping and Abduction', 'Dacoity', 'Robbery', 'Arson', 'Hurt',
       'Prevention of atrocities (POA) Act',
       'Protection of Civil Rights (PCR) Act', 'Other Crimes Against SCs'].sum().reset_index()

fig = go.Figure()
fig.add_trace(go.Scatter(x= year['Year'], y= year['Murder'],
                    name='Murder',line=dict(color='pink', width=4)))
fig.add_trace(go.Scatter(x= year['Year'], y= year['Rape'],
                    name='Rape',line=dict(color='green', width=4)))
fig.add_trace(go.Scatter(x= year['Year'], y= year['Kidnapping and Abduction'],
                    name='Kidnapping and Abduction',line=dict(color='orange', width=4))
             )
fig.add_trace(go.Scatter(x= year['Year'], y= year['Dacoity'],
                    name='Dacoity',line=dict(color='yellow', width=4)))
fig.add_trace(go.Scatter(x= year['Year'], y= year['Robbery'],
                    name='Robbery',line=dict(color='black', width=4)))
fig.add_trace(go.Scatter(x= year['Year'], y= year['Arson'],
                    name='Arson',line=dict(color='skyblue', width=4)))
fig.add_trace(go.Scatter(x= year['Year'], y= year['Hurt'],
                    name='Hurt',line=dict(color='royalblue', width=4)))
fig.add_trace(go.Scatter(x= year['Year'], y= year['Prevention of atrocities (POA) Act'],
                    name='Atrocities',line=dict(color='firebrick', width=4)))
fig.add_trace(go.Scatter(x= year['Year'], y= year['Protection of Civil Rights (PCR) Act'],
                    mode='lines+markers',
                    name='Civil Rights Violations'))
fig.add_trace(go.Scatter(x= year['Year'], y= year['Other Crimes Against SCs'],
                    name='Other Crimes',line=dict(color='red', width=4)))

fig.update_layout(uniformtext_minsize= 20,
    title_text="Total Crimes Against Scheduled castes 2001-2013",
   )
    
fig.show()
```

```python
stateyr = crime[crime.DISTRICT == 'Total']
stateyr = stateyr.groupby(['Year','STATE/UT'])['Murder', 'Rape',
       'Kidnapping and Abduction', 'Dacoity', 'Robbery', 'Arson', 'Hurt',
       'Prevention of atrocities (POA) Act',
       'Protection of Civil Rights (PCR) Act', 'Other Crimes Against SCs'].sum().reset_index()
```

```python
stateyr['sum'] =  stateyr.iloc[:, 2:].sum(axis=1)
```

```python
stateyr2 = stateyr.groupby('STATE/UT')['sum'].sum().reset_index()
```

```python
stateyr2 = stateyr2.sort_values('sum', ascending = False)
```

```python
import seaborn as sns
plt.figure(figsize = (20,20))
sns.set_context("talk")
plt.style.use("fivethirtyeight")
ax = sns.barplot(x = 'sum', y = 'STATE/UT', data = stateyr2)
plt.title('Total crimes against SCs')
for p in ax.patches:
        ax.annotate("%.f" % p.get_width(), xy=(p.get_width(), p.get_y()+p.get_height()/2),
            xytext=(5, 0), textcoords='offset points', ha="left", va="center")
```

```python
#import dependencies
crimestate = crime[crime.DISTRICT == 'Total']
crimestate =crimestate.groupby(['STATE/UT'])['Murder', 'Rape',
       'Kidnapping and Abduction', 'Dacoity', 'Robbery', 'Arson', 'Hurt',
       'Prevention of atrocities (POA) Act',
       'Protection of Civil Rights (PCR) Act', 'Other Crimes Against SCs'].sum().reset_index()

crimestate = crimestate[(crimestate.Murder > 100) & (crimestate.Rape > 100)]
plt.figure(figsize=(20,30))
columns = ['Murder', 'Rape',
       'Kidnapping and Abduction', 'Dacoity', 'Robbery', 'Arson', 'Hurt',
       'Prevention of atrocities (POA) Act',
       'Protection of Civil Rights (PCR) Act', 'Other Crimes Against SCs']
for i,column in enumerate(columns):
    crimestate = crimestate.sort_values(column,ascending = False)
    plt.subplot(5,2,i+1)
    ax = sns.barplot(data= crimestate,x= column ,y='STATE/UT')
    plt.xlabel('')
    plt.ylabel('')
    plt.title(column,size = 20)
    for p in ax.patches:
        ax.annotate("%.f" % p.get_width(), xy=(p.get_width(), p.get_y()+p.get_height()/2),
            xytext=(5, 0), textcoords='offset points', ha="left", va="center")
        
plt.tight_layout()
plt.subplots_adjust(hspace=.3)
plt.show()

crimestate['sum']=crimestate.sum(axis=1)
```

```python
scd = crime[crime.DISTRICT != 'Total']
scd = scd.groupby(['DISTRICT'])['Murder', 'Rape',
       'Kidnapping and Abduction', 'Dacoity', 'Robbery', 'Arson', 'Hurt',
       'Prevention of atrocities (POA) Act',
       'Protection of Civil Rights (PCR) Act', 'Other Crimes Against SCs'].sum().reset_index()
plt.figure(figsize=(20,40))

for i,column in enumerate(columns):
    scd1 = scd.sort_values(column,ascending = False)
    scd1 = scd1.head(10)
    plt.subplot(5,2,i+1)
    ax= sns.barplot(data= scd1,x= column ,y='DISTRICT' )
    plt.xlabel('')
    plt.ylabel('')
    plt.title(column,size = 20)
    for p in ax.patches:
        ax.annotate("%.f" % p.get_width(), xy=(p.get_width(), p.get_y()+p.get_height()/2),
            xytext=(5, 0), textcoords='offset points', ha="left", va="center")
         
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```
